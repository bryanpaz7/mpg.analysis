```python
#import seaborn
import seaborn as sns
#naming data as mpg to use in the code
mpg = sns.load_dataset('mpg')
#import matplotlib
import matplotlib.pyplot as plt
#list the first five entries for reference
mpg.head()
```

### Create a visual showing the relationship between horsepower and fuel economy.

```python
sns.scatterplot(data=mpg,
                x="horsepower",
                y="mpg",)
plt.show()
```

```python
## The scatterplot illustrates a clear inverse relationship between horsepower and miles per gallon (mpg), showing that as horsepower
## increases, fuel efficiency decreases.
```

### Create a table summarizing fuel economy by country of origin.

```python
sns.displot(data=mpg, x="mpg", hue="origin", col="origin",
            legend=False)
plt.show()
```

```python
## The distrubution shows that vehicles from the USA tend to have lower mpg, with a concentration around 15-25 mpg. In contrast,
## Japanese and European cars generally achieve higher fuel efficiency, with most Japanese cars ranging between 25-40 mpg and Eurpoean
## cars peaking around 25-35 mpg.
```

### Create a visual showing how fuel economy has changed over time.

```python
sns.scatterplot(data=mpg, y="mpg", x="model_year", hue="mpg")

plt.show()
```

```python
## The trend indicates an overall increase in fuel efficiency from 1970 to 1982, as more vehicles achieve higher mpg values in later years.
## This suggests advancements in automotive technology and possibly regulatory changes that promoted better fuel efficiency.
```

### Create two other visuals or tables related to fuel economy.

```python
sns.lineplot(data=mpg,
                x="mpg",
                y="weight",)
plt.gca().invert_yaxis()
plt.show()
```

```python
sns.displot(data=mpg,
            x="mpg",
            hue="cylinders",
            multiple="stack")
plt.show()
```

```python
## For both visualizations, it's clear that weight and fuel economy are inversely related, with heavier vehicles generally consuming more fuel.
## The number of cylinders in a vehicle also impacts fuel economy, with cars having fewer cylinders (typically smaller engines) tending to be
## more fuel-efficient.
```
